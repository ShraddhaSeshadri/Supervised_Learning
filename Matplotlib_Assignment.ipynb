```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
cars_df = pd.read_csv('mtcars.csv')
```

```python
cars_df.columns
```

**1. Generate a line plot graph for the column ‘model’ and ‘hp’.**
1. Map the ‘model’ column on the x-axis.
2. Map the ‘hp’ column on the y-axis.
3. Provide the x-axis label as Models of the cars.
4. Provide the y-axis label as Horse-Power of Cars.
5. Set the title as Model Names vs HorsePower.

```python
fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(cars_df['model'], cars_df['hp'])
plt.xticks(rotation=45)
plt.xlabel("Models of cars")
plt.ylabel(" Horse-Power of Cars")
plt.title("Model Names vs HorsePower.")
plt.show()
```

**2. Generate a bar plot graph for the columns ‘carbs’ and ‘gear’.**
1. Map the ‘carbs’ onto the x-axis.
2. Map the ‘gear’ onto the y-axis.
3. Provide the x-axis label as Number of carburetors.
4. Provide the y-axis label as Number of forward gears.
5. Set the title as carbs vs gear.

```python
import matplotlib.pyplot as plt

x = cars_df['carb']
y = cars_df['gear']

fig, ax = plt.subplots()
ax.bar(x, y)

# Add labels and title
ax.set_xlabel('Carburetors')
ax.set_ylabel('Number of Gears')
ax.set_title('Carburetors vs Gears')

# Show the chart
plt.show()
```

**3.Plot a histogram for the column ‘wt’.**
1. Map the ‘wt’ onto the x-axis.
2. Provide the x-axis label as ‘weight of the cars’.
3. Provide the y-axis label as ‘Count’
4. Set the number of bins as 30.
5. Set the title as ‘Histogram for the weight values’

```python
plt.hist(cars_df['wt'], bins=30)
plt.xlabel('weight of the cars')
plt.ylabel('Count')
plt.title('Histogram for the weight values')
plt.show()
```

**4. Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data frame.**

```python
fig, ax = plt.subplots()
explode = (0.1, 0, 0) 
ax.pie(cars_df['cyl'].value_counts(), labels=cars_df['cyl'].value_counts().index, explode=explode, autopct='%1.1f%%')
ax.set_title('Distribution of Cylinder Count')
plt.show()
```

**5. Plot the area chart for the colmns: ‘am’ and ‘carb’.**
1. Set the ‘am’ on the x-axis.
2. Set the ‘carb’ on the y-axis.
3. Provide the x-axis label as Transmission.
4. Provide the y-axis labe as Number of carburetors.
5. Provide the title as Transmission vs Number of carburetors

```python
mtcars = pd.read_csv('mtcars.csv')
grouped_df = mtcars.groupby(['am', 'carb'])['carb'].count().reset_index(name='count')
pivot_df = grouped_df.pivot(index='carb', columns='am', values='count').fillna(0)
pivot_df.plot.area()
plt.xlabel('Transmission')
plt.ylabel('Number of carburetors')
plt.title('Transmission vs Number of carburetors')
plt.show()
```
