```python
import pandas as pd

# read the data into a pandas dataframe
data = pd.read_csv('data.csv')
```

```python
data.columns
```

```python
# filter the rows for employees having more than 5 years experience and earning more than 60000
filtered_data = data[(data['YearsExperience'] > 5) & (data['Salary'] > 60000)]

# count the number of rows in the filtered data
count = len(filtered_data)

print("Number of employees having more than 5 years experience and earning more than 60000:", count)
```

```python
num_employees = len(data[(data["Salary"] >= 50000) & (data["Salary"] <= 80000)])
print("Number of employees earning between 50000-80000:", num_employees)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score


# split data into features and target
X = data.iloc[:, :-1].values
y = data.iloc[:, -1].values

# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)

# create the linear regression model
regressor = LinearRegression()
regressor.fit(X_train, y_train)

# predict the test set results
y_pred = regressor.predict(X_test)

# plot the test set results
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, y_pred, color='blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

# calculate the r2 score for the new model
r2 = r2_score(y_test, y_pred)
print('R2 score:', r2)
```

```python
data = pd.read_csv('housing.csv')
```

```python
data
```

```python
num_waterfront = len(data[data["waterfront"] == 1])
print("Number of houses with a waterfront:", num_waterfront)
```

```python
floors_counts = data['floors'].value_counts()
num_two_floors = floors_counts.loc[2.0]
print("Number of houses with 2 floors:", num_two_floors)
```

```python
old_houses = data[data['yr_built'] < 1960]

waterfront_count = len(old_houses[old_houses['waterfront'] == 1])

print("Number of houses built before 1960 with a waterfront:", waterfront_count)
```

```python
more_than_4_baths = data[data['bathrooms'] > 4]

most_expensive = more_than_4_baths['price'].max()

print("The price of the most expensive house having more than 4 bathrooms is:", most_expensive)
```
